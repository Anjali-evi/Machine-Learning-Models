### Hierarchial Clustering:

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score as sil

nc = [2,3,4,5,6,7,8,9]
sc_score = []
for i in nc:
    hc = AgglomerativeClustering(n_clusters=i, affinity='euclidean',linkage='complete')
    
    y = hc.fit_predict(dataset_scaled.iloc[:,:-1])
    sc_score.append(sil(dataset_scaled.iloc[:,:-1],y))
    
plt.plot(nc,sc_score,color="black")
plt.scatter(nc,sc_score,color="red")
plt.show()
```

```python
sc_score
```

```python
import scipy.cluster.hierarchy as sch

merge = sch.linkage(dataset_scaled, method = 'single')
cluster =  AgglomerativeClustering(n_clusters= 3,affinity='euclidean',linkage='single')
cluster.fit_predict(dataset_scaled)
y_pred = cluster.fit_predict(dataset_scaled)
dataset['hc'] = y_pred
```

### KMEANS Clustering:

```python
from sklearn.cluster import KMeans
```

```python
dataset_scaled
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(dataset_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
model_kmeans = KMeans(n_clusters= 2)
y_kmeans = model_kmeans.fit_predict(dataset_scaled)
```

```python
# converting labels to series from nd array
labels = pd.Series(model_kmeans.labels_)
```
