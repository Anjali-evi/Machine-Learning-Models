```python
movie_matrix = data_movie.pivot_table(index='userId' , columns='movie' , values='rating')
```

```python
movie_matrix
```

```python
movie_matrix = movie_matrix.fillna(0)
```

```python
movie_matrix
```

```python
# calculating cosine similarity
from sklearn.metrics import pairwise_distances  # pairwise dissimilarity
from scipy.spatial.distance import cosine, correlation
```

```python
user_similarity = 1 - pairwise_distances(movie_matrix.values, metric='cosine')
```

```python
user_similarity.shape
```

```python
# matrix returning similarity between all users 2182 * 2182
user_similarity
```

```python
user_sim = pd.DataFrame(user_similarity)
```

```python
user_sim
```

```python
# indexing the columns and rows of this user_similarity matrix as movie_matrix
user_sim.index = list(movie_matrix.index)
user_sim.columns = list(movie_matrix.index)
```

```python
user_sim.head(5)
```

```python
# we don't want to recommend movies already watched by the users
# therefore we will nullify the diagonal values from 1 -> 0
import numpy as np
np.fill_diagonal(user_similarity , 0)
```

```python
user_sim.sample(10)
```

### Creating recommendation:

```python
# here we take a particular user_id and we sort other users similar to the user_id
user_sim.sort_values([278836],ascending=False).head(100)
```

```python
def get_reco(customer_id):
    '''
    input is user_Id
    tem--> is going to contain all the customers who are similar to 8
    movie_list-->movies watched by top 100 similar customers to 8
    all set(movie_list) -> all the movies which are watched by 8
    '''
    tem = list(user_sim.sort_values([customer_id], ascending=False).head(100).index)
    movie_list = []
    for i in tem:
        movie_list = movie_list+ list(data_movie[data_movie["userId"]==i]["movie"])
    print(movie_list)
    return set(movie_list)-set(data_movie[data_movie["userId"]==customer_id]["movie"])
```

```python
get_reco(835)
```

```python
get_reco(276729) == 'Clara Callan'  # false
# here user_Id = 276729 has already watched movie Clara Callan; so the model don't suggest him that one
```
